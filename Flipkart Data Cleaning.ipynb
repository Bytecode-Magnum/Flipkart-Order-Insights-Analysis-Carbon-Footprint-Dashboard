```python
#: importing the datasets
import pandas as pd
df=pd.read_csv("Flipkart Order List.csv")
import warnings
warnings.filterwarnings('ignore');
for column in df.columns[:-1]:
    if df[column].dtype=='object':
        df[column]=df[column].str.upper()
```

```python
#: cleaning the customer name column
df['Customer_Name']=df['Customer_Name'].apply(lambda x: x.split(" ")[0])
print(df['Customer_Name'].value_counts())
```

```python
#: cleaning and fixing the date column
df['Date']=df['Date'].apply(lambda x: x.replace(",",""))
```

```python
#: fixing the date columns
df.loc[0:36,'Date']=df.loc[0:36]['Date']+" 2023"
mask = ((df.index >= 37) & (df.index <= 110) & ((df.Order_Status == 'REFUNDED') | (df.Order_Status == 'CANCELLED')))
df.loc[mask,'Date']=df.loc[mask,'Date']+' 2022'
mask = ((df.index >= 111) & (df.index <= 172) & ((df.Order_Status == 'REFUNDED') | (df.Order_Status == 'CANCELLED')))
df.loc[mask,'Date']=df.loc[mask,'Date']+' 2021'
mask = ((df.index >= 173) & (df.index <= 201) & ((df.Order_Status == 'REFUNDED') | (df.Order_Status == 'CANCELLED')))
df.loc[mask,'Date']=df.loc[mask,'Date']+' 2020'
mask = ((df.index >= 202)  & ((df.Order_Status == 'REFUNDED') | (df.Order_Status == 'CANCELLED')))
df.loc[mask,'Date']=df.loc[mask,'Date']+' 2019'
df['Date']=df['Date'].str.replace(r'^(MON|TUE|WED|THU|FRI|SAT|SUN)\s','',regex=True)
df['Date'] = df['Date'].str.replace(r'(\d+)(?:st|nd|rd|th)', r'\1', regex=True)
mask=((df.Order_Status == 'REFUNDED') | (df.Order_Status == 'CANCELLED'))
df.loc[mask,'Date']=df.loc[mask,'Date'].apply(lambda x: x.replace("TH "," ").replace("ND "," ").replace("RD "," ").replace("ST "," "))
```

```python
def change_format(date):
    dates=date.split(" ")
    return dates[1]+" "+dates[0]+" "+dates[-1]
df.loc[mask,'Date']=df.loc[mask,'Date'].apply(change_format)

```

```python
#: few orders dont have the current price so we take the price at which it was bought as the current price
df.loc[df.Current_Price.isnull(),'Current_Price']=df.loc[df.Current_Price.isnull()]['price']
```

```python
#$#: converting the date to date time column
df['Date']=pd.to_datetime(df['Date'],format='%b %d %Y')
```

```python
carbon_footprint_ranges = {
    'CLOTHING AND ACCESSORIES': (10, 30),
    'AUDIO & VIDEO': (10, 50),
    'BEAUTY AND GROOMING': (5, 20),
    'MOBILES & ACCESSORIES': (5, 15),
    'FOOTWEAR': (10, 30),
    'COMPUTERS': (20, 50),
    'HOME FURNISHING': (20, 60),
    'FOOD PRODUCTS': (5, 25),
    'FURNITURE': (20, 100),
    'HOME & KITCHEN': (15, 50),
    'KITCHEN, COOKWARE & SERVEWARE': (10, 30),
    'PENS & STATIONERY': (2, 5),
    'HEALTH CARE': (10, 30),
    'SPORTS': (5, 30),
    'AUTOMOTIVE ACCESSORIES': (10, 40),
    'HOME DECOR': (10, 40),
    'HEALTH & PERSONAL CARE APPLIANCES': (10, 30),
    'EXERCISE & FITNESS': (5, 20),
    'WEARABLE SMART DEVICES': (5, 20),
    'BAGS, WALLETS & BELTS': (5, 20),
    'FESTIVE DECOR & GIFTING': (2, 10),
    'BABY CARE': (5, 25),
    'KIDS ACCESSORIES': (5, 20)
}

# this is the average or range of carbon foot print generated by a category of a item
```

```python
def get_footprint(cat):
    value=carbon_footprint_ranges.get(cat)
    return (  int(value[0])+int(value[1]))/2
    
```

```python
df['average_carbon_footprints']=df['Main_category'].apply(get_footprint)
```
